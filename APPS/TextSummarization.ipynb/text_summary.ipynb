```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY", "")
grok_api_key = os.getenv("GROQ_API_KEY")
model_name = "Llama3-8b-8192"

llm = ChatGroq(model=model_name)
print(llm)
```

```python
speech = """On the one hand, [the iPhone has] been great, because now pretty much everyone in the world has a phone, and that’s 
kind of what enables pretty amazing things,” Zuckerberg said in an episode of the “Joe Rogan Experience.” 
“But on the other hand ... they have used that platform to put in place a lot of rules that I think feel arbitrary and 
[I] feel like they haven’t really invented anything great in a while. It’s like Steve Jobs invented the iPhone, 
and now they’re just kind of sitting on it 20 years later."""
```

```python
from langchain.schema import (
    AIMessage,
    HumanMessage,
    SystemMessage
)

chat_message = [
    SystemMessage(content='You are an export to summarize the speech.'),
    HumanMessage(content=f'please summarize the following speech: {speech}'),
]

tokenNumber = llm.get_num_tokens(speech)

print(f"Token Number: {tokenNumber}")
```

```python
llmResponse = llm(chat_message)
print("llm Response: ", llmResponse)
```

```python
## Secound way to summarize the speech
from langchain.chains import LLMChain
from langchain import PromptTemplate

genericTemplate = """ Write a summary of the following speech and translate it to {language}:
{speech}
"""
prompt = PromptTemplate(
    input_variables = ['language', 'speech'],
    template = genericTemplate
)

print("Raw Prompt: \n", prompt, "\n")

formattedPrompt = prompt.format(language='Turkish', speech=speech)

print("Formatted Prompt: \n", formattedPrompt, "\n")
tokenNumber = llm.get_num_tokens(formattedPrompt)

print(f"Token Number: {tokenNumber}")
```

```python
llmChain = prompt | llm

summary = llmChain.invoke({
    'language': 'Turkish',
    'speech': speech
})

print("Summary: \n", summary)
```

```python
# Methods
## Stuff Document Chain
## Map Reduce Chain
## Refine Chain
```

```python
## Stuff Document Chain
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("zucker.pdf")
pages = loader.load_and_split()
print(pages)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain_type = 'stuff'

template= """ Write a summary of the following speech:
{speech}
"""

prompt = PromptTemplate(
    input_variables = ['speech'],
    template = template
)

chain = load_summarize_chain(
    llm,
    chain_type=chain_type,
    prompt=prompt,
    verbose=True,
    document_variable_name='speech'  # Fix the validation error
)


output = chain.invoke(input={
    "input_documents": pages
})
```

```python
## Map Reduce Chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
splittedText = RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(pages)
print(splittedText)

print("length of splittedText: ", len(splittedText))
```

```python
template= """ Write a summary of the following speech:
{text}
"""

mapPromptTemplate = PromptTemplate(
    input_variables = ['text'],
    template = template
)

chain_type = 'map_reduce'

finalPrompt = """
Provide the final summary of the speech.
Add a motivational title and be kind to the reader.
speech: {text}
"""

finalPromptTemplate = PromptTemplate(
    input_variables = ['text'],
    template = finalPrompt
)

summary = load_summarize_chain(
    llm,
    chain_type=chain_type,
    map_prompt = mapPromptTemplate,
    combine_prompt = finalPromptTemplate,
    map_reduce_document_variable_name='text',
    verbose=True
)

output = summary.invoke(input={
    "input_documents": splittedText
})
```

```python
## Refine Chain
chain = load_summarize_chain(
    llm=llm,
    chain_type='refine',
    verbose=True
)
output = chain.invoke(input={
    "input_documents": splittedText
})
print(output)
```
