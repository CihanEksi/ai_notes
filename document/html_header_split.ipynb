```python
from langchain_text_splitters import HTMLHeaderTextSplitter
url = 'https://plato.stanford.edu/entries/reflective-equilibrium/'

headerToSplitOn = [
    ('h1','Header 1'),
    ('h2','Header 2'),
    ('h3','Header 3'),
]
html_splitter= HTMLHeaderTextSplitter(headerToSplitOn)
splits = html_splitter.split_text_from_url(url)
splits
```
